```python
import yfinance as yf
tsla = yf.Ticker("TSLA")
data1 = tsla.history(period="max")
tesla = pd.DataFrame(data1) 
tesla['ticker'] = 'TSLA'
tesla
```

```python

amaz = yf.Ticker("AMZN")

data2 = amaz.history(period="max")
amazon = pd.DataFrame(data2)
amazon['ticker'] = 'AMZN'
amazon
```

```python
finaldata = pd.concat([amazon,tesla], axis=0)
finaldata = finaldata.drop(['Dividends', 'Stock Splits'], axis = 1)
finaldata
#g = data.groupby(['ticker']).count()

finaldata.to_csv('/content/drive/MyDrive/stockprice.csv')
```
